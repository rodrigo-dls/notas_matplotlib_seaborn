## Subplot
Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una __matriz de gráficos__

```python
x = np.linspace(0,5,11)
y = x**2
```

Quiero trabajar con más de un gráfico

```python
plt.subplot(1,3,1) #argumentos representan: el nro de filas, el nro de columnas, el index que voy a manejar entre fil&col definido anteriormente
plt.plot(x,y, 'r--')
plt.subplot(1,3,2)
plt.hist(x,color='red')
plt.subplot(1,3,3)
plt.pie(x)
plt.show()
```

Colocando dos visualizaciones en un mismo gráfico

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b--')
plt.subplot(1,2,2)
plt.pie(x)
plt.show()
```

Creando visualizaciones en filas

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b--')
plt.subplot(2,1,2)
plt.pie(x)
plt.show()
```

Haciendo un grid amplio de varias visualzisaciones

```python
# Creo un grid de 6 bloques, es decir, 2 filas y 3 columnas
plt.subplot(2,3,1)
plt.plot(x,y ,'r--') # relación x con y
plt.plot(y,x) # relación y con x

plt.subplot(2,3,2)
plt.pie(x) # gráfico de torta

plt.subplot(2,3,3)
plt.bar(x,y) # gráfico de barras

plt.subplot(2,3,4)
plt.hist(x) # histograma de x

plt.subplot(2,3,5)
plt.scatter(x,y) # gráfico de distribución

plt.subplot(2,3,6)
plt.boxplot(x) # gráfico de cajas de x con y

plt.show()
```

## Método orientado a objetos
Es la alternativa a la visualizacion con 'pyplot' (rápido, fácil, díficil de customizar y entrar a detalles).
Este método es un poco más díficil y lleva más código de usar pero permite mayor personalización, más amigable y más a detalle. Puedo crear distintos tipos de gráficos y distribución en uno solo.

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure() # creo una figura y la asigno al objeto 'fig'. La figura representa el lienzo, espacio para visualizar
axes = fig.add_axes([0.1,0.1,0.5,0.9])  # cada axes representa un gráfico
axes.plot(x,y) # cada axes tiene sus propios ejes
fig.show() # el resultado es el mismo a pyplot hasta ahora, pero a partir de acá puedo avanzar más a detalle
```

Puedo visualizar dos gráficos en un mismo lienzo, cada objeto es independiente del otro y lo puedo customizar como quiera. Al mismo tiempo logro poner ambos en un mismo lienzo

```python
fig = plt.figure() # creo una figura y la asigno al objeto 'fig'
axes = fig.add_axes([0.1,0.1,0.8,0.9])  # los dos primeros parametros hacen referencia a la posición de la figura en el lienzo
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])  # los dos último parametros hacen referencia al ancho y alto del lienzo
axes.plot(x,y,'b')
axes.set_facecolor('pink')
axes_2.plot(y,x,'r')
axes_2.set_facecolor('gray')
fig.show() 
```

## Subplots
Trabajar múltiples gráficas en un solo lienzo con el método `orientado a objetos`

```python
x = np.linspace(0,5,21)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

Creo más gráficos(`axes`) y accedo a ellos a través de un índice, ya que la respuesta de `subplots` para los `axes` es de tipo `ndarrray`

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

Otra forma de atacar el problema con sintaxis más amigable, utilizando una tupla

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4) # creo un canvas de 8 axes
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x),'y')
ax3.plot(x, np.tan(x),'r')
ax4.plot(x, np.cos(x)**2, 'g')
fig.tight_layout() # para dar un formato amigable al grid de axes
plt.rcParams["figure.figsize"] = (10, 7)
# Reto agregar gráfica a la segunda linea del fig

ax5.plot(x, x*2)
ax6.plot(x, -np.sin(x),'y')
ax7.plot(x, -np.tan(0.5+x), 'r')
ax8.plot(x, -np.cos(x+2), 'g')
plt.show()
```

## Leyendas, etiquetas, títulos, tamaño
Agrego contexto a los gáficos mejorando la visualización.

### Con el método orientado a objetos

- `figsize` para ajustar el tamaño del plot
- `set_title` para dar título al gráfico
- `set_xlabel` & `set_ylabel` para dar título a los ejes
- `legend()` para imprimir el label del gráfico

```python
fig, axes = plt.subplots(1,2, figsize=(15,5)) # figsize usa unidades en pulgadas (inches)
axes[0].plot(x,y, label="$sin(x)$") # usando el operador "$" se imprime como expresión matemática
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend() # imprime el label del axe

axes[1].plot(y,x, label="$sin(x)**(-1)$")
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].legend()

plt.show()
```

### Con Pyplot
Los parámetros son:
- argumento `label` para dar nombre a la curva
- `title` para dar titulo al gráfico
- `xlabel` & `ylabel` para dar nombre a los ejes
- `legend()` para imprimir el nombre de la curva

```python
plt.plot(x,y, label="sin(x)")
plt.title('Este es el título de la gráfica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left', bbox_to_anchor=(0.76,0.25)) # por default se coloca donde mejor quede pero se pueden usar los parámetros 'loc' y 'bbox_to_anchor' para ajustarlo
```

## Bar plot
permite graficar variables categóricas. Se suele usar el eje X para compararlas.

```python
countries = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300] # por millones de personas
```

```python
plt.bar(countries, population, width=0.5, color=['green', 'brown'], label='Países con alta población')
plt.title('Poblacion por páis')
plt.xticks(np.arange(5), ('India', 'Japon', 'Mejico', 'Colombia', 'Alemania'), rotation=45) # para manipular las caractiresticis de los marcadores del eje X
plt.ylabel('millones de personas')
plt.xlabel('Paises')
plt.legend(frameon=True)
plt.show()
```

### Grafico de barras apilado
Volver a llamar al método `.bar()` con tu nueva variable y añadirle el parámetro `bottom=previous_data`. Y si quieres añadir más información aún, tendrás que sumar toda la información previa: `botom = previous_data_1 + previous_data_2`

```python
paises = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
hombres = [600, 400, 800, 300, 100]
mujeres = [400, 400, 100, 700, 200]
poblacion = [h + m for h,m in zip(hombres,mujeres)]

plt.bar(paises, hombres, label='Hombres', color='C4', width=0.5)
plt.bar(paises, mujeres, label='Mujeres', bottom=hombres, color='C8', width=0.4)
plt.xticks(rotation=45, size=10)
plt.title('Hombres y Mujeres por páis')
plt.xlabel('Países', size=15)
plt.ylabel('Hombres y Mujeres')
plt.legend(frameon=True)
plt.figure(figsize=(1,1)) 
plt.show()
```

## Histograma
Representa la distribución de los valores dentro del dataset, por lo que sólo utiliza una variable para crearse.

```python
data = np.random.randint(1,50,100)
# print(data, '\n', len(data), 'datos random')
data2 = np.random.randint(1,50,100)
# print(data2, '\n', len(data2), 'datos random')
```

```python
plt.style.use('default') # uso estilo classic para ver mejor los resultados
plt.hist(data, bins=25, histtype='bar', color='lightpink') # bins representa el numero de barras en el cual va a llegar la distribución de datos
plt.title('Histograma 1')
plt.show()
```

```python
plt.hist(data, bins=15, histtype='step')
plt.title('Histograma de tipo `step`')
plt.show()
```

### Superponer histogramas
usando `alpha` para trasparencia

```python
plt.hist(data, bins=5, alpha= 0.7, color= 'lightpink', label='hist 1')
plt.hist(data2, bins=5, alpha= 0.3, color= 'green', label='hist 2')
plt.title('2 histogramas superpuestos')
plt.legend()
plt.show()
```

Con `density` el gráfico se trasformará a porcentajes para que el área total sumada de los rectángulos sea de 1 (100%).

```python
plt.hist(data, bins=5, alpha= 0.7, color= 'lightpink', density=True)
plt.hist(data2, bins=5, alpha= 0.3, color= 'green', density=True)
plt.title('aplicando `density` para transformar la escala a base 100')
plt.show()
```

## Boxplot

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True) # patch_artist rellena el box, permite apreciar más el rango intercuartílico
plt.show()
```

```python
data = np.append(data, 200) # agrego un outliers que se pasa de 100 (el maximo valor en `data`)
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.title('Boxplot contemplando outliers')
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.title('Boxplot descartando los outliers')
plt.show()
```
