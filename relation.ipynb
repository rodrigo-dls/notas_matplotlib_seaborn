```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Relation

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

## Scatterplot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', style='sex');
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.1,0.5))
```

```python
plt.figure(figsize=(8,8))
markers = { "Lunch":"D", "Dinner":"s"} # predefino los markers con un diccionario
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.1,0.5));
```

## Lineplot
Sólo sirve aplicar con mínimo de variables categóricas

```python
plt.figure(figsize=(8,8))
markers = { "Lunch":"D", "Dinner":"s"} # predefino los markers con un diccionario
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.1,0.5));
```

## [Relplot](https://seaborn.pydata.org/generated/seaborn.relplot.html)
Condensa los gráficos de `scatterplot` y `lineplot`, ajustando el parámetro `kind` (`scatter` es _default_)

```python
plt.figure(figsize=(10,10))
markers = { "Lunch":"D", "Dinner":"s"} # predefino los markers con un diccionario
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers, col='day', row='time');
```

```python
plt.figure(figsize=(8,8))
markers = { "Lunch":"D", "Dinner":"s"} # predefino los markers con un diccionario
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers, kind='line');
```
