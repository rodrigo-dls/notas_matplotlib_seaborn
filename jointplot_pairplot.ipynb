```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head(2)
```

# [Jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)
Me une dos o más gráficos en un sólo cuadro
Al cargar dos variables (x,y), por default, carga un diagrama de dispersión (scatter) y una distribución en los ejes (hist)

```python
sns.jointplot(data=tips, x='total_bill', y='tip');
```

Si defino la variable `hue`, Seaborn, automáticamente cambie el histograma de los ejes por gráficas _kde_

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex');
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde');
```

el parámetro `marginal_ticks` agrega ejes a los gráficos de los ejes

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
                 marginal_kws=dict(bins=25, fill=False, multiple='stack'));
```

# [Pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)
Crea un grid de gráficas relacionando diferentes variables. Carga _por default_ todas las variables numéricas.
En la diagonal, una variable con sí misma, carga un _histograma_ y en los cruces hace _scatterplots_.

```python
sns.pairplot(data=tips);
```

Si defino el parámetro `hue`, crea un _kde_ en la diagonal

```python
sns.pairplot(data=tips, hue='sex', corner=True, kind='reg');
```

```python
sns.pairplot(data=tips, hue='sex', corner=True, kind='scatter', diag_kind='hist', diag_kws=dict( multiple= 'stack', palette='dark'),
height=2, aspect=1.5, plot_kws=dict(marker="+", linewidth=1),);
```

The return object is the underlying `PairGrid`, which can be used to further customize the plot:

```python
g = sns.pairplot(tips, diag_kind="kde")
g.map_lower(sns.kdeplot, fill=True, alpha=0.5, levels=4, color=".2"); # puedo agregar gráficos extra sobre el grid resultante
```
