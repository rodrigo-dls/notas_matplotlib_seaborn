```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head(2)
```

# [Heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
gráfico enfocado a una estructura matricial. Heatmap correlaciona todas las variables numéricas del dataset.

`.corr()` funcionalidad de Pandas para encontrar la correlación entre variables. Esta es en el rango [-1;1], siendo:
- __1__ correlación directa
- __0__ no hay correlación
- __-1__ correlación indirecta

```python
tips.corr()
```

Si tuviera muchas variables, es tedioso para leer. Acá entra el heatmap y su utilidad.

```python
sns.heatmap(tips.corr());
```

- `annot` imprime los valores de correlación dentro de los cuadros
- `cmap` define la paleta de colores
- `linewidths` agrega espacio entre los cuadros del mapa
- `linecolor` define un color para el espacio entre cuadros
- `vmin` & `vmax` define el rango para la referencia del mapa de colores
- `cbar` visualiza la barra de referencia, _bool_ (default is `True`)
- `center` centra el mapa de colores a un valor específico

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='gray',
vmin=0.5, vmax=1, cbar=True);
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='YlGnBu', linewidths=5, linecolor='gray',
cbar=True);
```
